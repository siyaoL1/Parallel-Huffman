### Example with real image

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

def plot_hist(img):
    # Show image
    plt.imshow(img, cmap='gray')
    plt.show()

    img_array = np.array(img)
    img_flat = img_array.flatten()

    # Plot a histogram of the pixel values
    plt.hist(img_flat, bins=256, range=(0, 255))
    plt.title('Pixel Distribution')
    plt.xlabel('Pixel Value')
    plt.ylabel('Count')
    plt.show()

    print("Dimension of image: {}".format(img_array.shape))
    print("Number of pixels: {}".format(np.prod(img_array.shape)))

# Calculate entropy of image
def get_entropy(img):
    hist, bins = np.histogram(img, bins=256, range=(0, 256))
    p = hist / img_flat.size
    p = p[p > 0]
    entropy = -np.sum(p * np.log2(p))
    return round(entropy, 1)

img = Image.open('images/dog.jpeg')
plot_hist(img)
print("Entropy of image: {}".format(get_entropy(img_flat)))
```

### Hypothetical example images

```python
# Generate a random image (highest entropy)
img = np.random.randint(256, size=(500, 500, 3)).astype('uint8')
plot_hist(img)
print("Entropy of image: {}".format(get_entropy(img)))
# save to bmp
Image.fromarray(img).save('images/high_ent.bmp')

img = np.zeros((500, 500, 3), dtype='uint8')
img[:, :250] = 255
plot_hist(img)
print("Entropy of image: {}".format(get_entropy(img)))
Image.fromarray(img).save('images/low_ent.bmp')
```
